```python
import pandas as pd
import numpy as np

data = pd.read_csv("Dataset1 .csv")

print(data.head())
print(data.tail())
```

```python
print(data.isnull().sum())
```

```python
data = (data.drop_duplicates())
```

```python
data.columns = data.columns.str.strip().str.replace(" ", "_")
data.columns = data.columns.str.lower()

print(data.columns)
```

```python

data['departure_time'] = pd.to_datetime(data['departure_time'], errors='coerce')
data['arrival_time'] = pd.to_datetime(data['arrival_time'], errors='coerce')
```

```python
data['journey_duration_hours'] = (
    (data['arrival_time'] - data['departure_time']).dt.total_seconds() / 3600
)
```

```python
if 'stops' in data.columns:
    data['stops'] = data['stops'].fillna(0)
```

```python
data.loc[data['journey_duration_hours'] < 0, 'journey_duration_hours'] += 24

train_summary = data.groupby('train_no').agg(
    Total_Distance=('distance', 'sum'),
    Total_Stops=('station_name', 'count'),
    Avg_Journey_Duration_Hours=('journey_duration_hours', 'mean')
).reset_index()

print(train_summary.head())
```

```python
data_sorted = data.sort_values(['train_no', 'distance'])

source = data_sorted.groupby('train_no')['station_name'].first().value_counts().reset_index()
source.columns = ['station_name', 'Train_Count']

destination = data_sorted.groupby('train_no')['station_name'].last().value_counts().reset_index()
destination.columns = ['station_name', 'Train_Count']
```

```python
print('Source Station Traffic:\n')
print(source)
```

```python
print('Destination Station Traffic:\n')
print(destination)
```

```python
train_sz = data.groupby('train_no').agg(
    source_station=('station_name', 'first'),
    destination_station=('station_name', 'last'),
).reset_index()

print(train_sz.head())
```

```python
data.to_csv("clean_train_data.csv", index=False)
train_summary.to_csv("train_summary.csv", index=False)
train_sz.to_csv("train_source_destination.csv", index=False)
```
